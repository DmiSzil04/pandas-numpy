**Задание 3**\
Найдите самое популярное имя среди пассажиров:
+ мужчин;
+ женщин.

1. Разобьём столбец name, поскольку нам нужно имя, а там склеено (по разделителю ".", а затем по разделителю "("), уберём значения, которые могут в дальнейшем запутать

```python
df['only_name'] = df['name'].str.split('.').apply(lambda x: x[-1])
df['only_name'] = df['only_name'].str.split('(').apply(lambda x: x[0])
df['only_name'] = df['only_name'].str.split(' ').apply(lambda x: x[1:])
df['only_name'] = df['only_name'].apply(lambda x: [i for i in x if '"' not in i])
df['only_name'] = df['only_name'].apply(lambda x: [i for i in x if len(i) > 2])
df['only_name']
```

2. Создадим списки с полом, склеенным именем, в которых значения повторяются столько раз, сколько имён имеет человек

```python
sex_list = []
for i in range(len(df['sex'])):
    for j in range(len(df['only_name'][i])):
        sex_list.append(df['sex'][i])

sex_list
```

```python
an_list = []
for i in range(len(df['name'])):
    for j in range(len(df['only_name'][i])):
        an_list.append(df['name'][i])

an_list
```

3. Создадим кортеж из пары пол-имя, в котором множественные имена разделены

```python
name_list = list(zip(sex_list, [x for y in df['only_name'] for x in y]))
name_list
```

4. Из имеющихся данных создадим временный датасет

```python
temp_df = pd.DataFrame(name_list, columns=['sex', 'only_name'])
temp_df['name'] = pd.Series(an_list)
temp_df
```

```python
temp_df.value_counts(subset=['only_name', 'sex'])
```

5. Мне показалось странным, что среди женщин лидирует имя William, я решил дополнительно проверить на ошибки
